```python
import pandas as pd

csv1 = pd.read_csv('/Users/rasmuskopperudriis/Coding/work_folder/runfolder/INF055_results/mutation/app_merged_mutation_vaccine.csv')
csv2 = pd.read_csv('/Users/rasmuskopperudriis/Coding/work_folder/runfolder/INF055_summary.csv')


csv1_HA = csv1[csv1['Ref_Name'].str.contains('HA')]
csv1_NA = csv1[csv1['Ref_Name'].str.contains('NA')]

pivoted_df1_HA = csv1_HA.groupby('sample')['Differences'].apply(';'.join).reset_index().rename(columns={'Differences': 'Vaccine Mutations HA'})
pivoted_df1_NA = csv1_NA.groupby('sample')['Differences'].apply(';'.join).reset_index().rename(columns={'Differences': 'Vaccine Mutations NA'})

merged_df = pd.merge(pivoted_df1_HA, pivoted_df1_NA, on='sample', how='outer')
merged_df = merged_df.rename(columns={'sample': 'Sample'})



final_merge = pd.merge(csv2, merged_df, on='Sample', how='outer')

final_merge.head()




#pivoted_csv1 = csv1.groupby(['sample','Ref_Name'])['Differences'].apply(';'.join).reset_index()
#pivoted_csv1.head()







```
